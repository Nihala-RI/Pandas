# Group by and Aggregating

```python
import pandas as pd
```

```python
df = pd.read_csv(r"E:\Self-learning\Pandas\Flavors (1).csv")
df
```

```python
df.groupby('Base Flavor')
```

```python
df.groupby('Base Flavor')['Flavor Rating'].mean()
```

```python
df.groupby('Base Flavor').mean()
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').sum(numeric_only=True)
```

```python
df.groupby('Base Flavor').sum(numeric_only=True).loc[['Chocolate']] #similar to having clause in SQL
```

```python
df.groupby('Base Flavor')['Flavor rating'].mean()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':'mean','Texture Rating':['mean','sum']})
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','sum'],'Total Rating':['count'],'Texture Rating':['median']})
```

```python
list(df.groupby('Base Flavor')['Flavor Rating'])
```

```python
df.groupby('Base Flavor')['Flavor Rating'].agg(lambda x: x.max() - x.min())
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':lambda x: x.max() - x.min()})
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','sum']})
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum','min'], 'Texture Rating':['mean','max','count','sum'] })
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating': ['mean','max','count','sum']})
```
